```python
df["Difference"] = df["Price_in_thousands"] - df["resale_value"]
```

```python
df.head(50)
```

```python
df_resale = df.dropna()[["Manufacturer","Difference"]]
```

```python
df_resale = df_resale.groupby("Manufacturer").mean()
```

```python
df_resale.reset_index(inplace = True)
```

```python
df_resale
```

```python
df
```

```python
df = pd.merge(df,df_resale,how = "inner",on = "Manufacturer")
```

```python
df.drop(columns = ["Difference_x"],inplace = True)
```

```python
df.rename(columns = {"Difference_y":"Average Diffrerence"},inplace = True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
def replace_resale(row):
    if pd.isnull(row["resale_value"]):
        return row["Price_in_thousands"] - row["Average Diffrerence"]
    else:
        return row["resale_value"]
```

# Apply function

```python
df["resale_value"] = df.apply(replace_resale,axis = 1)
```
