```python
users_dict = dict(zip(indexes,list(range(76353))))
```

```python
songs_dict = dict(zip(cols,list(range(10000))))
```

```python
data = np.zeros([len(df_users["user_id"].unique()),len(df_users["song_id"].unique())])
```

```python
for i in range(df_users.shape[0]):
    user_id = df_users.iloc[i,0]
    user_index = users_dict[user_id]
    song_id = df_users.iloc[i,1]
    song_index = songs_dict[song_id]
    listen_count = df_users.iloc[i,2]
    data[user_index,song_index] = listen_count
    if i%10000 == 0:
        print(f"Rows completed {i}")
```

```python
df = pd.DataFrame(data = data,index = list(users_dict.keys()), columns = list(songs_dict.keys()))
```
