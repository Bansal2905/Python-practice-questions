# Does name have anything to do with survival of passenger

```python
def fetch_title(name):
    index1 = name.index(",")
    index2 = name.index(".")
    return name[index1+2:index2]
```

```python
df["title"] = df["Name"].apply(fetch_title)
```

```python
df
```
